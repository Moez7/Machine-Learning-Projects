```python
# Import the dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
nba = pd.read_csv('nba_2013.csv')
nba.head(7)
```

```python
nba.shape
```

```python
#Find the average value for each numerical culumn
nba.mean()
```

```python
# Get the mean / average of specific column 
nba.loc[:, 'fg'].mean()
```

```python
#Make a pairwise scatter plot 
sns.pairplot(nba[["ast", "fg", "trb"]])
```

```python
#Make Heat map to see correlations
correlation = nba[["ast", "fg", "trb"]].corr()
sns.heatmap(correlation, annot= True)
```

```python
# Make the cluster of players using kMeans
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=5, random_state = 1) #Creating KMeans model with 5 clusters
good_columns = nba._get_numeric_data().dropna(axis=1) #get numeric data and remove columns with missing data  
kmeans_model.fit(good_columns) #train the model
labels = kmeans_model.labels_ #Get the labels or (cluster label for each player)
labels
```

```python
# Plot players by cluster
from sklearn.decomposition import PCA 
pca_2 = PCA(2)
plot_columns = pca_2.fit_transform(good_columns)
plt.scatter(x=plot_columns[:,0], y = plot_columns[:,1], c= labels)
plt.show()
```

```python
#Find player LeBron
LeBron = good_columns.loc[ nba['player'] == 'LeBron James',: ]

#Find player Durant
Durant =good_columns.loc[ nba['player'] == 'Kevin Durant',: ]
print(LeBron)
print(Durant)
```

```python
# Converting the data into lists for model prediction
LeBron_list = LeBron.values.tolist()
Durant_list = Durant.values.tolist()

# The cluster classification of LeBron James and Kevin Durant
LeBron_cluster = kmeans_model.predict(LeBron_list) #LeBron belongs to cluster 3
Durant_cluster = kmeans_model.predict(Durant_list) #LeBron belongs to cluster 3

#print the cluster label
print(LeBron_cluster)
print(Durant_cluster)
```

```python
#Split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(nba[['fg']], nba[['ast']], test_size = 0.2, random_state=2)
```

```python
# Predict the number of assist per player from the field goals made
#Create the Linear Regression Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train) #training the model
predictions = lr.predict(x_test) #Make predictions on the test data
print(predictions) #print the predictions
print(y_test) # Print the actual values
```

```python
#Test Model: Score returns the coefficient of determination R^2 of the prediction
lr_confidence = lr.score(x_test, y_test)
print("lr confidence:" , lr_confidence)
```
